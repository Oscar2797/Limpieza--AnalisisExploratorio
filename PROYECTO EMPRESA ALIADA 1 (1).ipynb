```python
# AGRUPAR ITEM CON SUMAS DE TOTAL_UNIT_SALES, TOTAL_VALUE_SALES Y TOTAL_UNIT_AVG_WEEKLY_SALES

FACT_SALES2=FACT_SALES[["TOTAL_UNIT_SALES","TOTAL_VALUE_SALES", "TOTAL_UNIT_AVG_WEEKLY_SALES","ITEM"]].groupby("ITEM").sum()
FACT_SALES2
```

```python
# DESHACER ASIGNACIÓN DE ÍNDICES FACT_SALES2

FACT_SALES2= FACT_SALES2.reset_index()
FACT_SALES2.head()
```

```python
# JUNTAR DIM_PRODUCT CON SUMAS DE FACT_SALES2 POR ITEM

DIM_PRODUCT2= pd.merge(DIM_PRODUCT, FACT_SALES2[["ITEM","TOTAL_UNIT_SALES","TOTAL_VALUE_SALES","TOTAL_UNIT_AVG_WEEKLY_SALES"]], how="left")
DIM_PRODUCT2.head()
```

```python
# JUNTAR DIM_PRODUCT2 CON SEGMENTO DE DIM_SEGMENT

DIM_PRODUCT4= pd.merge(DIM_PRODUCT2, DIM_SEGMENT[["FORMAT","ATTR1","ATTR2","ATTR3","SEGMENT"]], how="left")
DIM_PRODUCT4.head()
```

```python
# INFORMACIÓN DIM_PRODUCT3

DIM_PRODUCT4.info()
```

```python
#RELLENAR VALORES NULOS CON CEROS DIM_PRODUCT4

DIM_PRODUCT4.fillna(0,inplace=True)
DIM_PRODUCT4.info()
```

```python
# CAMBIAR NOMBRE DE COLUMNA "CATEGORY" DE DIM_PRODUT4 PARA HOMOLOGAR CON DATA FRAME DE DIM_CATEGORY

DIM_PRODUCT4.rename(columns={"CATEGORY":"ID_CATEGORY"},inplace=True)
DIM_PRODUCT4.head()
```

```python
# JUNTAR DIM_PRODUCT4 CON CATEGORY DE DIM_CATEGORY

DIM_PRODUCT5= pd.merge(DIM_PRODUCT4, DIM_CATEGORY[["ID_CATEGORY","CATEGORY"]], how="left")
DIM_PRODUCT5.head()
```

```python
# AGRUPAR WEEK CON SUMAS DE TOTAL_UNIT_SALES, TOTAL_VALUE_SALES Y TOTAL_UNIT_AVG_WEEKLY_SALES

FACT_SALES3=FACT_SALES[["TOTAL_UNIT_SALES","TOTAL_VALUE_SALES", "TOTAL_UNIT_AVG_WEEKLY_SALES","WEEK"]].groupby("WEEK").sum()
FACT_SALES3.head()
```
