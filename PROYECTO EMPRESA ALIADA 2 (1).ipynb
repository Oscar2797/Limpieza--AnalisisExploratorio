```python
# CAMBIAR NOMBRE DE COLUMNA "ITEM_CODE" DE FACT_SALES PARA HOMOLOGAR CON DATA FRAME DE DIM_PRODUCT5

FACT_SALES.rename(columns={"ITEM_CODE":"ITEM"},inplace=True)
FACT_SALES.head()
```

```python
# JUNTAR FACT_SALES CON SEGMENTO DE DIM_PRODUCT5

FACT_SALES= pd.merge(FACT_SALES, DIM_PRODUCT5[["ITEM","SEGMENT"]], how="left")
FACT_SALES.head()
```

```python
# TIPOS DE SEGMENTOS DISTINTOS EN DATA FRAME FACT_SALES
FACT_SALES["SEGMENT"].unique()
```

```python
# FUNCIÓN PARA GENERAR TABLAS DE SUMAS DE TOTAL_UNIT_SALES,	TOTAL_VALUE_SALES Y	TOTAL_UNIT_AVG_WEEKLY_SALES POR SEMANA DE CADA SEGMENTO
segmentos=['POWDER', 'PRETREAT', 'BAR', 'LIQUID & GEL', 'BLEACH', 'SANITIZER','OTHERS']
def tabla_seg(segmento):
    segmento= FACT_SALES[FACT_SALES["SEGMENT"]==segmento]
    segmento=segmento[["TOTAL_UNIT_SALES","TOTAL_VALUE_SALES", "TOTAL_UNIT_AVG_WEEKLY_SALES","WEEK"]].groupby("WEEK").sum()
    segmento= segmento.reset_index()
    return segmento
```

```python
POWDER_SEGMENT=tabla_seg("POWDER")
PRETREAT_SEGMENT=tabla_seg("PRETREAT")
BAR_SEGMENT=tabla_seg("BAR")
LIQUID_GEL_SEGMENT=tabla_seg("LIQUID & GEL")
BLEACH_SEGMENT=tabla_seg("BLEACH")
SANITIZER_SEGMENT=tabla_seg("SANITIZER")
OTHERS_SEGMENT=tabla_seg("OTHERS")
```

```python
# FUNCIÓN PARA GENERAR GRÁFICOS DE LÍNEAS

def graficas_de_lineas(ejes,dataframe,título,color):

    ejes.plot(dataframe["WEEK"],dataframe["TOTAL_VALUE_SALES"],label=título, color= color)
    ejes.axes.xaxis.set_ticks([])
    ax.set_ylabel("VENTAS TOTALES")
    ax.set_xlabel("DATOS SEMANALES :01-22 - 52-23")
    plt.title(título)
```

```python
# FUNCIÓN PARA GENERAR TABLAS DE SUMAS DE TOTAL_UNIT_SALES,	TOTAL_VALUE_SALES Y	TOTAL_UNIT_AVG_WEEKLY_SALES POR SEMANA DE CADA REGIÓN
regiones=['TOTAL AUTOS AREA 5', 'TOTAL AUTOS AREA 6','TOTAL AUTOS SCANNING MEXTICO', 'TOTAL AUTOS AREA 3', 'TOTAL AUTOS AREA 2', 'TOTAL AUTOS AREA 1', 'TOTAL AUTOS AREA 4']
def tabla_reg(region):
    region= FACT_SALES[FACT_SALES["REGION"]==region]
    region=region[["TOTAL_UNIT_SALES","TOTAL_VALUE_SALES", "TOTAL_UNIT_AVG_WEEKLY_SALES","WEEK"]].groupby("WEEK").sum()
    region= region.reset_index()
    return region
```
